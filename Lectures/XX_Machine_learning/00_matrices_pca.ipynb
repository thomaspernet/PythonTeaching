# Inverse Matrix

```
import numpy as np
```

```
a  = np.array([
    [4, -7],
    [2, -3]
    ])
a_1 = np.linalg.inv(a) 
a_1
```

When you multiply the original matrix A times the inverse A –1 , the result is the 2 × 2 identity matrix.

```
np.dot(a , a_1)
```

## Matrix not singular

If the number ad – bc (obtained by finding the difference of the cross-products) is equal to 0, then you’d have to divide by 0. That’s an impossible situation — no numbers have 0 in their denominator.

```
b = np.array([
    [8, 2],
    [12, 3]
    ])
try:
  np.linalg.inv(b)
except:
  print('Singular matrix')
```

## PCA

```
# principal component analysis 
from numpy import array 
from numpy import mean 
from numpy import cov 
from numpy.linalg import eig 
```

```
A = array([ 
    [1, 2],
    [3, 4],
    [5, 6]]) 
```

```
# column means (by column)
M = mean(A.T, axis=1) 
# center columns by subtracting column means 
C = A - M 
# calculate covariance matrix of centered matrix 
V = cov(C.T) 
print(M, '\n', C, '\n', V)
```

```
# factorize covariance matrix 
values, vectors = eig(V) 
print(vectors)
print(values)
```

```
vectors.T
```

```
# project data 
P = vectors.T.dot(C.T)
print(P.T) 
```

```
# Principal Component Analysis
from numpy import array
from sklearn.decomposition import PCA
X = np.array(
    [
    [7 , 4 , 3],
    [4 , 1 , 8],
    [6 , 3 , 5],
    [8 , 6 , 1],
    [8 , 5 , 7],
    [7 , 2 , 9],
    [5 , 3 , 3],
    [9 , 5 , 8],
    [7 , 4 , 5],
    [8 , 2 , 2]
]
)
# create the PCA instance
pca = PCA(3)
# fit on data
pca.fit(X)
# access values and vectors
print(pca.components_)
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)  
# transform data
B = pca.transform(X)
print(B)
```

## Real Example

```
import pandas as pd
df = pd.read_csv("http://r-marketing.r-forge.r-project.org/data/rintro-chapter8.csv")
df.head()
```

### Variables meaning

| Perceptual adjective (column name) | Example survey text             |
|------------------------------------|---------------------------------|
| perform                            | Brand has strong performance    |

| latest                             | Brand has the latest products   |
| fun                                | Brand is fun                    |

| value                              | Brand products are a good value |

```
df_normalized = df.iloc[:,:-1].apply(lambda x: (x - x.mean())/x.std(), axis=0)
df_normalized.describe()
```

Correlation plot for the simulated consumer brand ratings. This visualization of the basic data appears to show three general clusters that comprise:

- fun/latest/trendy
- rebuy/bargain/value
- perform/leader/serious, respectively

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Generate a large random dataset
rs = np.random.RandomState(33)

# Compute the correlation matrix
corr = df_normalized.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

“What is the average (mean) position of the brand on each adjective?”

Brands f and g are similar—with high ratings for rebuy and value but low ratings for latest and fun. 

Other groups of similar brands are b/c, i/h/d,a n da/

```
df_norm = pd.concat([df_normalized, df['brand']], axis = 1)
df_ag_brand = df_norm.groupby('brand').mean()
sns.heatmap(df_ag_brand)
```

### PCA

```
pca = PCA(9)
# fit on data
pc = pca.fit_transform(df_normalized)
# access values and vectors
print(pca.components_)
#print(pca.explained_variance_)
#print(pca.explained_variance_ratio_)  
```

transform the Data using the first 2 components

```
pc_df = pd.DataFrame(data = pc, 
        columns = ['PC1', 'PC2','PC3','PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9'])
pc_df['brand'] = df['brand']
pc_df.head()
```

where adjectives map in four regions: 

- category leadership (“serious,” “leader,” and “perform” in the upper right)
- value (“rebuy,” “value,” and “bargain”), trendiness (“trendy” and “latest”),
- finally “fun” on its own. 

```
fig, ax = plt.subplots(figsize = (14,8))
feature_vectors = pca.components_.T
arrow_size, text_pos = 7.0, 8.0
ax = sns.scatterplot(x="PC1", y="PC2",data=pc_df)
for i, v in enumerate(feature_vectors):
    ax.arrow(0, 0, arrow_size*v[0], arrow_size*v[1], head_width=0.2,
             head_length=0.2, linewidth=2, color='red')
    ax.text(v[0] * text_pos, v[1] * text_pos, list(df_normalized)[i],
            color='black', ha='center', va='center', fontsize=18)
```
