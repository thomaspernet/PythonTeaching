## Scatterplot

```python
# Import dataset 
midwest = pd.read_csv("midwest_filter.csv")
midwest.head()
```

```python
sns.scatterplot(x="area", y="poptotal",data=midwest)
```

```python
sns.scatterplot(x="area",
                y="poptotal",
                hue="category",
                data=midwest).legend(loc='center left',
                                     bbox_to_anchor=(1.25, 0.5),
                                     ncol=1)
```
